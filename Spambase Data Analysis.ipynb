```python
def outlier_remover(df, column_names):
    for column in column_names:
        
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
     
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return filtered_df

columns_to_clean = ['capital_run_length_average', 
               'capital_run_length_longest', 
               'capital_run_length_total']
df_cleaned = outlier_remover(data, columns_to_clean)
```

```python
# Plotting heat map
capital_tot_variable = df_cleaned[['capital_run_length_total']]
word_variables = df_cleaned.iloc[:,:48]

combined_df = pd.concat([capital_tot_variable, word_variables], axis=1)

corr_matrix = combined_df.corr()

filtered_corr = corr_matrix.loc[capital_tot_variable.columns, word_variables.columns]

plt.figure(figsize=(20, 8))
sns.heatmap(filtered_corr, cmap='coolwarm', center=0)
plt.title('Correlation between Total Capital Variable and Word Variables')
plt.yticks(rotation = 90)
plt.show()
```
